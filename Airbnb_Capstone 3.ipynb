## Latar Belakang Perusahaan

Airbnb adalah perusahaan teknologi global yang bergerak di bidang penyewaan akomodasi jangka pendek berbasis platform digital. Melalui situs web dan aplikasi mobile, Airbnb mempertemukan pemilik properti (host) dengan wisatawan atau penyewa (guest) yang mencari tempat tinggal sementara — seperti apartemen, rumah, atau kamar pribadi.

Di Bangkok, Airbnb berperan penting dalam ekosistem pariwisata. Kota ini menjadi salah satu destinasi paling populer di Asia Tenggara, baik untuk wisata maupun perjalanan bisnis. Airbnb hadir untuk:
* Menyediakan alternatif akomodasi yang lebih fleksibel dibanding hotel,
* Mendukung ekonomi lokal, karena banyak warga membuka unit pribadi mereka untuk disewa,
* Memberikan pengalaman lokal yang autentik bagi wisatawan yang ingin tinggal di lingkungan non-komersial.

Airbnb eksis karena menjawab kebutuhan yang tidak sepenuhnya dipenuhi oleh industri hotel tradisional.
Di konteks Bangkok, kebutuhan ini meliputi:
- Fleksibilitas dan Pilihan
    Wisatawan membutuhkan berbagai tipe akomodasi — dari apartemen murah hingga vila mewah. Airbnb memungkinkan penyewa untuk memilih sesuai kebutuhan, lokasi, dan anggaran mereka.
- Harga Lebih Kompetitif
    Airbnb sering kali menawarkan harga lebih terjangkau dibanding hotel, terutama untuk perjalanan panjang. Ini menjadikannya pilihan populer bagi digital nomad, backpacker, dan keluarga.
- Pengalaman Lokal
    Banyak tamu ingin “hidup seperti penduduk lokal”. Dengan menginap di hunian warga Bangkok (misalnya di Vadhana, Pathum Wan, atau Huai Khwang), mereka bisa menikmati budaya dan gaya hidup lokal yang tidak mereka dapatkan di hotel.
- Sumber Penghasilan untuk Warga
    Bagi tuan rumah lokal, Airbnb menyediakan cara mudah untuk menghasilkan pendapatan tambahan dari properti yang tidak terpakai.
 
 Hal ini penting di Bangkok, di mana sektor properti tumbuh cepat dan banyak unit apartemen kosong.

Airbnb tidak memiliki properti sendiri — seluruh model bisnisnya berbasis komisi dan layanan digital. Sumber pendapatan utama di Bangkok (dan global) meliputi:
- Komisi dari Pemilik Properti (Host Fee)
    Airbnb mengenakan biaya sekitar 3–15% dari setiap transaksi kepada host untuk penggunaan platform, manajemen pemesanan, dan dukungan pelanggan.
- Biaya dari Tamu (Guest Service Fee)
    Tamu juga dikenakan biaya layanan 10–14% di atas harga sewa, yang menjadi pendapatan langsung bagi Airbnb.
    Biaya ini digunakan untuk pemeliharaan platform, asuransi, dan dukungan pelanggan.
- Layanan Tambahan (Airbnb Experiences dan Partner Program) 

Di Bangkok, Airbnb juga memperoleh pendapatan dari:
- Airbnb Experiences – aktivitas tur lokal seperti kelas masak Thai atau tur kuil oleh penduduk lokal.
- Kerja sama dengan properti profesional / manajemen apartemen, yang menyediakan banyak unit sekaligus (co-hosting dan business hosting).

## Data Cleaning

##### Menghapus Duplikasi (drop_duplicates)

Langkah ini dilakukan untuk memastikan setiap listing unik hanya muncul satu kali dalam dataset.

* Kolom `id` digunakan sebagai identifier unik untuk setiap properti di Airbnb.
* Jika satu `id` muncul lebih dari sekali, artinya terjadi duplikasi data, yang bisa berasal dari proses scraping berulang atau kesalahan penggabungan dataset.
* Jika duplikasi tidak dihapus, maka:
    * Perhitungan rata-rata harga dan okupansi akan menjadi bias, karena satu properti dihitung lebih dari sekali.
    * Visualisasi seperti sebaran harga dan peta lokasi juga akan terdistorsi oleh data ganda.

Dapat disimpulkan bahwa data tidak ada yang duplikat karena jumlah data unik dari kolom `id`= jumlah total baris

#### Deteksi outlier di kolom `price`
Alasan bisnis: harga ekstrem (terlalu murah/mahal) dapat mendistorsi rata-rata dan strategi pricing.
Metode IQR:
- IQR = Q3 - Q1
- Lower bound = Q1 - 1.5 x IQR
- Upper bound = Q3 + 1.5 x IQR

```python
# === Deteksi Outlier Sederhana (IQR) untuk kolom harga ===
def iqr_outlier_bounds(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return lower, upper

if "price" in df.columns:
    lower_p, upper_p = iqr_outlier_bounds(df["price"].dropna())
    print(f"Batas outlier (IQR) untuk price: lower={lower_p:.2f}, upper={upper_p:.2f}")
    print("Jumlah outlier (di bawah lower atau di atas upper):",
        ((df["price"] < lower_p) | (df["price"] > upper_p)).sum())
```

Menghitung occupancy_rate (Proxy Tingkat Okupansi)

Airbnb tidak menyediakan kolom langsung untuk occupancy rate (tingkat keterisian kamar).
Oleh karena itu, kita membuat variabel proxy (pendekatan) berdasarkan kolom `availability_365`.
* `availability_365` menunjukkan berapa hari dalam setahun listing tersebut masih tersedia untuk disewa.
* Jika sebuah listing sering disewa, maka nilai `availability_365` akan rendah karena banyak tanggal sudah terisi.
* Sebaliknya, jika `availability_365` tinggi, berarti listing jarang disewa atau sering diblokir oleh host.

Dengan rumus:

occupancy_rate=(365−availability_365)/365​


Hasilnya berupa nilai antara 0 dan 1:

* 0 → listing tidak pernah disewa
* 1 → listing terisi penuh sepanjang tahun

Langkah ini penting karena:
* `occupancy_rate` menjadi indikator utama performa listing, menggambarkan tingkat permintaan dan revenue potensial.
* Dapat digunakan untuk analisis hubungan harga dengan okupansi, serta identifikasi area dengan tingkat sewa tertinggi.

```python
df["occupancy_rate"] = (365 - df["availability_365"]) / 365
df.head(5)
```

#### Revenue Index = Average Price × Average Occupancy Rate

```python
# Hitung rata-rata harga dan okupansi per tipe kamar
revenue_potential = df.groupby("room_type")[["price", "occupancy_rate"]].mean().reset_index()

# Tambahkan kolom Revenue Index
revenue_potential["revenue_index"] = revenue_potential["price"] * revenue_potential["occupancy_rate"]

# Urutkan berdasarkan potensi revenue tertinggi
revenue_potential = revenue_potential.sort_values("revenue_index", ascending=False)
print(revenue_potential)

# Visualisasi
plt.figure(figsize=(10,6))
plt.bar(revenue_potential["room_type"], revenue_potential["revenue_index"], color="red")
plt.title("Revenue Potential (Price × Occupancy Rate) by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Revenue Index (Price × Occupancy Rate)")
plt.show()
```

## Simulasi: Dampak Rekomendasi terhadap Revenue & Okupansi

Tujuan simulasi ini adalah untuk mengukur potensi peningkatan revenue dan tingkat okupansi (occupancy rate) apabila rekomendasi yang dihasilkan dari analisis sebelumnya diterapkan.

Kita akan mengasumsikan:
* Peningkatan harga dilakukan di wilayah dengan okupansi tinggi (misalnya Huai Khwang, Vadhana, dan Samphanthawong),
* Optimalisasi listing dan strategi harga menghasilkan kenaikan okupansi rata-rata 5–10%,
* Tidak terjadi trade-off besar antara kenaikan harga dan penurunan okupansi (karena kenaikan harga diimbangi oleh peningkatan kualitas atau promosi).

1. Asumsi Simulasi

| Asumsi                                | Nilai                          |
| ------------------------------------- | ------------------------------ |
| Kenaikan harga (price increase)       | +10%                           |
| Peningkatan okupansi (occupancy_rate) | +7% (rata-rata konservatif)    |
| Periode analisis                      | 1 tahun                        |
| Variabel revenue dasar                | `price × occupancy_rate × 365` |


```python
# Salin data dasar
sim_df = df.copy()

# Hitung revenue awal (baseline)
sim_df["revenue_baseline"] = sim_df["price"] * sim_df["occupancy_rate"] * 365

# Terapkan rekomendasi (kenaikan harga & okupansi)
sim_df["price_new"] = sim_df["price"] * 1.10  # kenaikan 10%
sim_df["occupancy_new"] = sim_df["occupancy_rate"] * 1.07  # kenaikan 7%
sim_df["occupancy_new"] = sim_df["occupancy_new"].clip(upper=1.0)

# Hitung revenue baru
sim_df["revenue_new"] = sim_df["price_new"] * sim_df["occupancy_new"] * 365

# Hitung persentase perubahan revenue
sim_df["revenue_change_%"] = ((sim_df["revenue_new"] - sim_df["revenue_baseline"]) / sim_df["revenue_baseline"]) * 100

# --- Tampilkan ringkasan perbandingan sebelum dan sesudah ---
print("=== Ringkasan Sebelum dan Sesudah Rekomendasi ===")
print(sim_df[["price", "occupancy_rate", "revenue_baseline", 
        "price_new", "occupancy_new", "revenue_new", 
        "revenue_change_%"]].head(10))  # tampilkan 10 baris pertama

# --- Hitung rata-rata perubahan ---
avg_change = sim_df["revenue_change_%"].mean()
avg_occ_change = (sim_df["occupancy_new"].mean() - sim_df["occupancy_rate"].mean()) * 100

print("\n=== Ringkasan Agregat ===")
print(f"Rata-rata peningkatan revenue: {avg_change:.2f}%")
print(f"Rata-rata peningkatan okupansi: {avg_occ_change:.2f}%")

# --- Visualisasi perbandingan sebelum vs sesudah ---
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.bar(["Before Recommendation", "After Recommendation"],
        [sim_df["revenue_baseline"].mean(), sim_df["revenue_new"].mean()],
        color=["gray", "green"])
plt.title("Average Revenue per Listing: Before vs After Recommendation")
plt.ylabel("Average Annual Revenue")
plt.show()

sim_dfgrp = sim_df.groupby("neighbourhood")[["revenue_baseline", "revenue_new"]].mean().reset_index()
```

Berdasarkan pola data Bangkok dan hasil analisis sebelumnya, perkiraan dampaknya adalah:

| Metrik                              | Sebelum | Sesudah | Perubahan (Riil) | Perubahan (Simulasi) |
| ----------------------------------- | ------- | ------- | ---------------- | -------------------- |
| Rata-rata harga (price)             | 1.600   | 1.760   | +10.00%          | +10.00%              |
| Rata-rata okupansi (occupancy_rate) | 0.36    | 0.39    | +1.92%           | +7.00%               |
| Rata-rata revenue tahunan           | 210.000 | 246.000 | **+17.16%**      | **+19.00%**          |


Jika host menerapkan strategi yang disarankan (penyesuaian harga, peningkatan layanan, dan fokus di area dengan okupansi tinggi), maka potensi peningkatan pendapatan tahunan mencapai ±19% per listing.

#### Insight dari simulasi

1. Kenaikan harga moderat (10%) di area dengan okupansi tinggi tidak menurunkan permintaan, justru meningkatkan margin keuntungan.
2. Optimalisasi okupansi (7%) dapat dicapai melalui:
    * Foto dan deskripsi listing yang lebih menarik,
